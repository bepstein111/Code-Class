#### Bin added a function to add Temperatures
__A:__ Because the default \__add\__ ("+") method is not defined/overrided when we writing the Temperature class. See the revised version in the following cell where you _can_ add two temperatures together.

```python
class UpdateTemperature(object): 

    def __init__(self,value):   
        self._celsius = float(value) 
        
    # define/override default __add__ method
    def __add__(self, other):
        return self._celsius + other._celsius
    
    def c(self):
        return str(self._celsius)+ ' C'   
    def f(self):
        return str((self._celsius * 9.0/5.0)+32.0) + ' F'
    
boil = UpdateTemperature(100)
freeze = UpdateTemperature(0)
print "boil + freeze = " + str(boil + freeze)
```

```python
#Ryan L
"""It's a basic calculator. It doesn't do anything spectacular 
(python can already do all of this in itself) but it would be a 
good exercise for new programmers to play around with to figure out 
how if, elif, and else statements work. It could be a cool challenge 
to see if they can figure out how to make it less error prone, too.

"""
def basic_calc(num1,oper,num2):
    if oper == "+":
        return num1 + num2
    elif oper == "-":
        return num1 - num2
    elif oper == "*" or oper == "x":
        return num1 * num2
    elif oper == "/":
        return num1 / num2
    else:
        return "Error"
    
basic_calc(1,"+",2)
```

```python
#Chance
"""My program is a Zenoian supertask which theoretically can
complete an infinite number of tasks in a finite amount of time, 
which is of course impossible in reality. The user can define a
larger minimum meaningful length than standard to make it possible.  """
import time
def Zeno():
    d = raw_input("Enter a distance: ")
    print "You have selected a distance of " + str(d) + " Meters."
    print "Now define the smallest meaningful distance. You may simply input \"Planck Length\" for your current dimension."
    i = raw_input("Smallest meaningful distance: ")
    if i[0] == "P" or i[0] == "p":
        i = 1.6 * 10** -35
    else:
        i = int(i)
    d = int(d)
    print "Attempting to travel " + str(d) + " Meters approaching a minimum distance of " + str(i) + "..."
    time.sleep(30)
    t = 30
    while d > i:
        t = t - ( t * 0.5)
        d = d - (d * 0.5)
        print "Traveled " + str(d) + " Meters..."
        time.sleep(t)
    if i == 1.6 * 10** -35:
        print "Supertask complete; infinity breached."
    else:
        print "Relative Supertask complete."
        
Zeno()
```

```python
#Bin
"""__A:__ For the best or the worst, it is easy to be dazzled by 
the idea of so-called "computational realization." The caveat 
lies on the pompous romanticization and premature philosophization
of this idea--people using it to create a mystical 
techno-futurist aura without really explaining 
what _is_ computational realization. And I think it is imperative 
for pedagogue to address this topic in a serious way. Here I 
present a simple example that can be used to help elicit much 
thinking from starter/novice about the idea of using 
computational method to realize a practical task.

I present to starters with the task of _swapping_, and code of 
how to achieve it. After reading the code, one should find that 
the seemingly facile task--swapping one number with the other--
suddenly becomes not so straightforward. The starter will see 
that swapping process actually involves: 
1. note down the value of the one number; 
2. assign the value of the other number to (the position of) the one; 
3. assign the noted-down value to (the position of) the other one. 

This function will not only give starter a better understanding 
of data type, memory position, but also let starter realizes the 
difference between the visual realization of task (imagining the 
graphical process of swapping) and computational realization 
(placing value in different positions). Plus this function can also serve as a segue into the introduction of simple sorting algorithm like bubble sort and more general topics like computational complexity.
"""
def swapping(seq, index1, index2):
    temporary = seq[index1]
    seq[index1] = seq[index2]
    seq[index2] = temporary
    
numberList = [0,1,2,3,4,6,5,7,8,9]
swapping(numberList, 5, 6)
print numberList
        
```
